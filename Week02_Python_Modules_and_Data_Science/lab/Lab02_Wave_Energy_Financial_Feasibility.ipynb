## Hints

- **LCOE formula:**

$$\text{LCOE} = \frac{\text{Total Discounted Costs}}{\text{Total Discounted Energy}} = \frac{\text{CAPEX} + \sum_{t=1}^{N} \frac{\text{OPEX}_t}{(1+r)^t}}{\sum_{t=1}^{N} \frac{E_t}{(1+r)^t}}$$

  where $r$ is the discount rate, $N$ is the project lifetime, and $E_t$ is annual energy production in MWh.

- **Annual energy production:** $E = \text{Capacity (kW)} \times \text{Capacity Factor} \times 8760 \text{ hours/year}$

- The dataset gives power in **Watts**. Be careful with unit conversions.

## 1. Wave Energy Analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt

# Load the WEC Perth dataset
df_wec = pd.read_csv('WEC_Perth_49.csv')

# Display basic information about the dataset
print("Dataset shape:", df_wec.shape)
print("\nFirst few rows:")
print(df_wec.head())
print("\nColumn names:")
print(df_wec.columns.tolist())
print("\nData types:")
print(df_wec.dtypes)
print("\nBasic statistics:")
print(df_wec.describe())
```

## 2. Competing Technologies

```python
# Compute statistics on farm power output
mean_power = df_wec['Total_Power'].mean()
min_power = df_wec['Total_Power'].min()
max_power = df_wec['Total_Power'].max()
std_power = df_wec['Total_Power'].std()

print(f"Mean farm power: {mean_power:.2f} W")
print(f"Min farm power: {min_power:.2f} W")
print(f"Max farm power: {max_power:.2f} W")
print(f"Std farm power: {std_power:.2f} W")

# Assume rated power per WEC is 500 kW = 500,000 W
# Justification: Based on typical wave energy converters like the Pelamis device, which has a rated power of around 750 kW, but for Perth's conditions and the dataset, we assume 500 kW per device as a conservative estimate. Source: IRENA Wave Energy report.
rated_per_wec = 500000  # W
total_rated_power = 49 * rated_per_wec  # W
capacity_factor_wave = mean_power / total_rated_power

print(f"Assumed rated power per WEC: {rated_per_wec / 1000:.0f} kW")
print(f"Total rated power for farm: {total_rated_power / 1000000:.1f} MW")
print(f"Estimated capacity factor for wave energy in Perth: {capacity_factor_wave:.3f}")
```

## 3. Financial Comparison

```python
# Define technologies data
# Sources: IRENA Renewable Power Generation Costs in 2022 for CAPEX and OPEX.
# Capacity factors: For Perth, Australia - Solar PV: CSIRO data ~22%, Onshore Wind: NREL ~35%, Wave: calculated from dataset.
# Lifetimes: IRENA estimates.

technologies = {
    'Wave Energy': {
        'capex': 7000,  # USD/kW
        'opex': 200,    # USD/kW/year
        'cf': capacity_factor_wave,
        'lifetime': 20,  # years
        'source': 'IRENA 2022 (emerging technologies), CF from dataset analysis'
    },
    'Solar PV': {
        'capex': 1200,  # USD/kW
        'opex': 20,     # USD/kW/year
        'cf': 0.22,
        'lifetime': 25,  # years
        'source': 'IRENA 2022, CF: CSIRO for Perth region'
    },
    'Onshore Wind': {
        'capex': 1500,  # USD/kW
        'opex': 40,     # USD/kW/year
        'cf': 0.35,
        'lifetime': 25,  # years
        'source': 'IRENA 2022, CF: NREL for Western Australia'
    }
}

# Display the data
for tech, data in technologies.items():
    print(f"{tech}:")
    print(f"  CAPEX: ${data['capex']}/kW")
    print(f"  OPEX: ${data['opex']}/kW/year")
    print(f"  Capacity Factor: {data['cf']:.3f}")
    print(f"  Lifetime: {data['lifetime']} years")
    print(f"  Source: {data['source']}")
    print()
```

## 4. Visualization

```python
import numpy as np

# Discount rate: 5% - typical for renewable energy projects in Australia, reflecting moderate risk and inflation. Source: Based on Australian Treasury guidelines.
discount_rate = 0.05

# Project capacity: 100 MW as per the requirement
capacity_mw = 100  # MW
capacity_kw = capacity_mw * 1000  # kW

def calculate_lcoe(capex, opex, cf, lifetime, capacity_kw, discount_rate):
    """
    Calculate Levelized Cost of Energy (LCOE) in USD/MWh
    """
    # Annual energy production in MWh
    annual_energy_mwh = capacity_kw * cf * 8760 / 1000
    
    # Annuity factor
    annuity = (1 - (1 + discount_rate) ** -lifetime) / discount_rate
    
    # Discounted costs
    discounted_capex = capex * capacity_kw
    discounted_opex = opex * capacity_kw * annuity
    
    # Discounted energy
    discounted_energy = annual_energy_mwh * annuity
    
    lcoe = (discounted_capex + discounted_opex) / discounted_energy
    return lcoe

def calculate_npv(capex, opex, cf, lifetime, capacity_kw, discount_rate, electricity_price=0.08):
    """
    Calculate Net Present Value (NPV) in USD
    Assuming electricity price of 0.08 USD/kWh (approximate Australian wholesale price)
    """
    npv = -capex * capacity_kw
    annual_energy_mwh = capacity_kw * cf * 8760 / 1000
    annual_revenue = annual_energy_mwh * electricity_price * 1000  # USD
    annual_cost = opex * capacity_kw
    
    for t in range(1, lifetime + 1):
        npv += (annual_revenue - annual_cost) / (1 + discount_rate) ** t
    
    return npv

# Calculate LCOE and NPV for each technology
results = {}
for tech, data in technologies.items():
    lcoe = calculate_lcoe(data['capex'], data['opex'], data['cf'], data['lifetime'], capacity_kw, discount_rate)
    npv = calculate_npv(data['capex'], data['opex'], data['cf'], data['lifetime'], capacity_kw, discount_rate)
    results[tech] = {'LCOE': lcoe, 'NPV': npv}

# Display results
print(f"Discount rate: {discount_rate * 100}%")
print(f"Project capacity: {capacity_mw} MW")
print(f"Assumed electricity price: $0.08/kWh")
print()
for tech, res in results.items():
    print(f"{tech}:")
    print(f"  LCOE: ${res['LCOE']:.4f}/MWh")
    print(f"  NPV: ${res['NPV']/1e6:.1f} million")
    print()
```

## 5. Recommendation

Based on the financial analysis, Perth should prioritize investing in **Onshore Wind** for the 100 MW renewable energy capacity addition, as it offers the lowest LCOE at approximately $48/MWh compared to $55/MWh for solar PV and $541/MWh for wave energy. The NPV for onshore wind is also the highest at $139 million, indicating strong profitability. Solar PV follows as a close second with competitive costs.

Wave energy, while promising for its predictability and minimal visual impact, is currently not competitive due to high capital costs and lower capacity factor derived from the dataset. It could become viable if technological advancements reduce CAPEX by 80% or more, or if policy incentives like feed-in tariffs specifically target emerging technologies. Additionally, wave energy might be considered in a diversified portfolio to balance intermittency of solar and wind.

The model does not capture environmental benefits, grid integration costs, land use impacts, or potential future cost reductions through learning curves. It also assumes constant capacity factors and costs, without accounting for degradation over time or regional variations in resource availability.

```python
# Plot 1: LCOE Comparison
tech_names = list(results.keys())
lcoe_values = [results[tech]['LCOE'] for tech in tech_names]

plt.figure(figsize=(8, 5))
plt.bar(tech_names, lcoe_values, color=['blue', 'green', 'red'])
plt.title('Levelized Cost of Energy (LCOE) Comparison', fontsize=14)
plt.ylabel('LCOE (USD/MWh)', fontsize=12)
plt.xlabel('Technology', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plot 2: CAPEX vs Capacity Factor
capex_values = [technologies[tech]['capex'] for tech in tech_names]
cf_values = [technologies[tech]['cf'] for tech in tech_names]

plt.figure(figsize=(8, 5))
plt.scatter(cf_values, capex_values, s=100, color=['blue', 'green', 'red'])
for i, tech in enumerate(tech_names):
    plt.annotate(tech, (cf_values[i], capex_values[i]), xytext=(5, 5), textcoords='offset points', fontsize=10)
plt.title('CAPEX vs Capacity Factor', fontsize=14)
plt.xlabel('Capacity Factor', fontsize=12)
plt.ylabel('CAPEX (USD/kW)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
